# From Convolutional Neural Networks (CNN) to Graph Neural Networks (GNN)

## Convolution operation
Given an image(matrix) a learnable kernel is defined, then an element wise multiplication is performed between the kernel and a submatrix from the image starting at (0,0), then the elemenst from the multiplication are added and inserted into the convolutional matrix.
image.png

Benefits of Convolution are:
* Parameter sharing
* Translation Invariant
* num of parameters in the learnable kernel is independent of input
Challenges in Generalizing Convolution for graphs:
* Number of neighbor nodes changes
* Distance between nodes changes
* Number of attributes can vary(features)
* We may have Heterogenous Graphs
* Node ordering can change (homorphisim problem)
## Signal processing point of view
Graphs are generic of signal structure, the spacing between samples(nodes) can be irregular.By introducing structure(edges) we provide additional information. CNN scale to large dimensions making them a good fit to grok with graph data, however CNN only work with regular structure

## Mesage Passing
The message pasing framework consist in the following:
* Initialization: initial state $h_v = x_v$
* Agregation: how other nodes influenve a given node $a_v=f_a({h_u|u\in N(v)})$
* Update: update the current node base on $a_v$ given by $h_v = f_u(a_u,h_v)$
